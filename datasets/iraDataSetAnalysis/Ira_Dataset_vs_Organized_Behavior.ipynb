```python
from sklearn.ensemble import RandomForestClassifier
#clf = RandomForestClassifier(n_estimators=100, random_state=0,class_weight={0:3,1:1})
clf = RandomForestClassifier(random_state=50,class_weight={0:3,1:1})
clf.fit(data1.drop('ClassId',axis=1), data1['ClassId'])
```

```python
testData = pd.read_csv(data_path+'IraTestDataFriendFollowerRatios.csv')
```

```python
testData['prediction'] = clf.predict(testData.drop("hashtag",axis=1))
```

```python
testData['ClassId'] = 1
```

```python
testData['prediction'].sum()
```

```python
from sklearn.metrics import precision_recall_fscore_support as score
precision, recall, fscore, support = score(testData['ClassId'], testData['prediction'])

print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
print('fscore: {}'.format(fscore))
print('support: {}'.format(support))
```
